## Introduccion teorica:

## Sistemas LTI (Lineales e Invariantes en el Tiempo)

Un **sistema** en procesamiento digital de señales es un conjunto de reglas, operaciones o transformaciones \(T\{\}\) que actúan sobre una secuencia de entrada \(x[n]\) para producir una secuencia de salida \(y[n]\):

$$
y[n] = T\{x[n]\}
$$

---

### Propiedades de un sistema LTI

Un sistema se dice **LTI** si cumple dos propiedades fundamentales:

---

### 1. Linealidad

Un sistema es **lineal** si respeta las propiedades de:

**Aditividad:**

$$
T\{x_1[n] + x_2[n]\} = T\{x_1[n]\} + T\{x_2[n]\} = y_1[n] + y_2[n]
$$

**Escalamiento (homogeneidad):**

$$
T\{a \cdot x[n]\} = a \cdot T\{x[n]\}
$$

> En palabras: la respuesta a una combinación lineal de entradas es la misma combinación lineal de las respuestas.

---

### 2. Invariancia en el tiempo

Un sistema es **invariante en el tiempo** si su salida **no depende del instante absoluto** en que se aplica la entrada.

Formalmente, si

$$
y[n] = T\{x[n]\}
$$

entonces para cualquier desplazamiento \(n_0\):

$$
T\{x[n - n_0]\} = y[n - n_0]
$$

> Esto significa que un desplazamiento en la entrada produce el mismo desplazamiento en la salida, sin cambiar su forma.



## Tipos de sistemas

Existen **dos tipos de sistemas**:

- **Sistemas de respuesta finita al impulso (FIR):**  
  Tienen respuestas al impulso de **duración finita en el tiempo**.

- **Sistemas de respuesta infinita al impulso (IIR):**  
  Tienen respuestas al impulso de **duración infinita en el tiempo**.



### Sistemas FIR

La relación general entrada–salida para un sistema FIR se expresa como:

$$
y[n] = \sum_{k=M}^{N} b_k \, x[n-k] \tag{2}
$$

donde $b_k$, $k$, $N$ y $M$ son constantes especificadas.  

La respuesta de este sistema a un impulso es fácil de calcular: simplemente se sustituye $x[n] = \delta[n]$ en la ecuación anterior, lo que por definición hace que la salida sea igual a $h[n]$:

$$
h[n] = \sum_{k=M}^{N} b_k \, \delta[n-k]
$$

Esto deja claro que la respuesta al impulso de un sistema FIR contiene un número finito de impulsos distribuidos en un rango limitado de tiempo, es decir, $M \leq n \leq N$.


## Sistemas IIR (Respuesta Infinita al Impulso)

La relación general **entrada–salida** para un sistema IIR causal se define mediante la **ecuación en diferencias lineal de coeficientes constantes (LCCDE)**:

$$
\sum_{k=0}^{N} a_k \, y[n-k] = \sum_{k=0}^{M} b_k \, x[n-k] \tag{2.3}
$$

Sin pérdida de generalidad, sea \(a_0 = 1\). Reescribiendo esta ecuación para despejar la salida \(y[n]\), obtenemos:

$$
y[n] = \sum_{k=0}^{M} b_k \, x[n-k] - \sum_{k=1}^{N} a_k \, y[n-k] \tag{2.4}
$$

Comparando la ecuación anterior con la de un sistema FIR, se observa la **principal diferencia** entre FIR e IIR:

- Para un **sistema FIR**, la salida depende **solo de valores presentes y pasados de la entrada**.  
- Para un **sistema IIR**, la salida depende **también de valores pasados de la salida**.


## Análisis y desarrollo:

### Sistema LTI IIR

La ecuación en diferencias del sistema es:

$$
y[n] = 0.03 x[n] + 0.05 x[n-1] + 0.03 x[n-2] + 1.5 y[n-1] - 0.5 y[n-2]
$$


Esta es una **ecuación en diferencias lineal de coeficientes constantes (LCCDE)**, lo que significa que el sistema es **lineal, invariante en el tiempo (LTI)** y de **respuesta infinita al impulso (IIR)**.


### 1.Graficar la señal de salida para cada una de las señales de entrada que generó en el TS1. Considere que las mismas son causales.

#### Parametros 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from IPython.display import display, Math
# ----------------------
# Parámetros
# ----------------------
N = 100        # número de muestras
f = 2000        # frecuencia de la señal
fs = 50000      # frecuencia de muestreo
fsquare = 4000  # frecuencia cuadrada
n = np.arange(N)  # eje discreto
t = np.arange(N) / fs


# Tiempo de simulación total
T_sim = N / fs

display(Math(f'T_{{sim}} = \\frac{{N}}{{f_s}} = \\frac{{{N}}}{{{fs}}} = {T_sim} \\, \\mathrm{{s}}'))
```

#### Funciones

```python
def ecuacion_diferencias(x, b, a):
    """
    Implementa la ecuación en diferencias:
        y[n] = b[0]*x[n] + b[1]*x[n-1] + ... + a[0]*y[n-1] + a[1]*y[n-2] + ...
    
    x : array_like -> señal de entrada
    b : list -> coeficientes de entrada (x[n], x[n-1], ...)
    a : list -> coeficientes de realimentación (y[n-1], y[n-2], ...)
    """
    N = len(x)
    y = np.zeros(N)
    M = len(b)
    K = len(a)
    
    for n in range(N):
        # Parte de la entrada
        for i in range(M):
            if n-i >= 0:
                y[n] += b[i] * x[n-i]
        # Parte de la realimentación
        for j in range(1, K+1):
            if n-j >= 0:
                y[n] += a[j-1] * y[n-j]
    return y

def energia_potencia(y):
    E = np.sum(y**2)
    P = np.mean(y**2)
    return E, P
    
def plot_signal(n, y, title):
    plt.figure()
    plt.plot(n, y, marker='o', linestyle='-')
    plt.title(title)
    plt.xlabel("desplazamiento [n]")
    plt.ylabel("y[n]")
    plt.grid(True)
    plt.show()
```

```python
# Coeficientes
# ----------------------
a = [0.0]*11
b = [0.0]*11
b = [0.03, 0.05, 0.03]
a = [1.5, -0.5]
```

```python
# Señales
# ----------------------
s1 = np.sin(2 * np.pi * f * t)
s2 = 2 * np.sin(2 * np.pi * f * t + np.pi/2)
s3 = np.sin(2 * np.pi * (f/2) * t)
sq = signal.square(2 * np.pi * fsquare * t)
```

```python
# Salidas
# ----------------------
y_s1 = ecuacion_diferencias(s1, b, a)
y_s2 = ecuacion_diferencias(s2, b, a)
y_s3 = ecuacion_diferencias(s3, b, a)
y_sq = ecuacion_diferencias(sq, b, a)
```

```python
# Graficar salidas
plot_signal(n, y_s1, "Salida con entrada s1[n]")
plot_signal(n, y_s2, "Salida con entrada s2[n]")
plot_signal(n, y_s3, "Salida con entrada s3[n]")
plot_signal(n, y_sq, "Salida con entrada sq[n]")
```

#### Se puede observar que para estos valores la y[n] oscila para señales senoideales y para la cuadrada esta va oscilando incrementando su valor.

#### Calculo de energia y potencia

```python
for y, name in zip([y_s1, y_s2, y_s3, y_sq], ['s1', 's2', 's3', 'sq']):
    E, P = energia_potencia(y)
    print(f"{name}: Energía = {E:.4f}, Potencia = {P:.4f}")
```

#### Calculo de Energia y potencia.

#### Respuesta al impulso h[n]

##### Parametros: 

```python
delta = np.zeros(N)
delta[0] = 1
h = ecuacion_diferencias(delta, b, a)
plot_signal(n, h, "Respuesta al impulso h[n]")
```

#### Como su respuesta al impulso a lo largo de 100 muestras convergen a un numero distinto de cero es un sistema de respuesta infinita en el tiempo.

### Calculo de y[n] con la señal "s1" como entrada usando convolucion:

```python
ysc1_np = np.convolve(s1, h)[:N]
plt.figure()
plt.plot(n, y_s1, marker='o', linestyle='-', label='Ecuación diferencias')
plt.plot(n, ysc1_np, marker='x', linestyle='--', label='Convolución')
plt.title("Comparación salida s1[n]")
plt.xlabel("desplazamiento [n]")
plt.ylabel("y[n]")
plt.grid(True)
plt.legend()
```
