# Introduccion teorica:


Señal analogica: La variable independiente (normalmente el tiempo) es un número real no cuantizado. Esto significa que la señal tiene un valor medible en cada instante de tiempo. Ejemplo: la salida de un micrófono.

Señal digital: Es una señal discreta en el tiempo, es decir, su variable independiente está cuantizada: solo tiene valor en ciertos instantes (por ejemplo, índices enteros n=0,1,2,…). Se representa como una secuencia de números x[n], donde cada número corresponde al valor de la señal en un instante particular.

Sintetizar y graficar:

Importaciones:

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

Funciones:

```python
def mi_funcion_sen(vmax=1.0, dc=0.0, ff=1.0, ph=0.0, nn=1000, fs=1000.0):
    """
    Sine wave generator.

    vmax: peak amplitude [V]
    dc  : mean value [V]
    ff  : frequency [Hz]
    ph  : phase [rad]
    nn  : number of samples
    fs  : sampling frequency [Hz]
    """
    tt = np.arange(nn, dtype=float) / float(fs)             # time vector [s]
    xx = dc + vmax * np.sin(2.0 * np.pi * ff * tt + ph)     # samples
    return tt, xx

def verificar_ortogonalidad(x, y):
    prod = np.sum(x * y)
    return prod
```

Parametros:

```python
fs = 400000     # Frecuencia de muestreo 20 kHz
N  = 20000      # Número de muestras
f  = 2000      # Frecuencia de la señal (2 kHz)
fsquare = 4000  # Frecuencia de la señal cuadrada [Hz]
```

Señales:

```python
# Señal original
tt, x1 = mi_funcion_sen(vmax=1, dc=0, ff=f, ph=0, nn=N, fs=fs)

# Señal amplificada (2x) y desfazada π/2
tt, x2 = mi_funcion_sen(vmax=2, dc=0, ff=f, ph=np.pi/2, nn=N, fs=fs)

# Misma señal modulada por una de la mitad de frecuencia (1 kHz)
_, mod = mi_funcion_sen(vmax=1, dc=0, ff=f/2, ph=0, nn=N, fs=fs)
x3 = x1 * mod  # modulación AM

# Señal cuadrada usando scipy.signal.square()
sq = signal.square(2 * np.pi * fsquare * tt)

# --- Clipping al 75% del pico ---

x3_clip = np.clip(x3, -0.75, 0.75)

# Pulso rectangular de 10 ms (nivel 1, luego 0)
pulso = np.where(tt < 0.01, 1.0, 0.0)

```

```python
Ts = 1 / fs
print("Tiempo entre muestras (Ts) =", Ts, "segundos")
```

```python
# Energía de cada señal
E_x1 = np.sum(x1**2)
E_x2 = np.sum(x2**2)
E_sq = np.sum(sq**2)
E_pulso = np.sum(pulso**2)

print("Energía x1 =", E_x1, )
print("Energía x2 =", E_x2)
print("Energía cuadrada =", E_sq)
print("Energía pulso =", E_pulso)
```

```python
plt.figure()
plt.plot(tt, x1)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.title("Señal sinusoidal de 2 kHz (s1)")
plt.grid(True)

# --- Gráfico 2: señal amplificada y desfazada ---
plt.figure()
plt.plot(tt, x2, color="orange")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.title("Señal amplificada y desfazada π/2 (s2)")
plt.grid(True)

plt.figure()
plt.plot(tt, x3, color='green')
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.title("3) Señal modulada en amplitud por 1 kHz (s3)")
plt.grid(True)

plt.figure()
plt.plot(tt, x3_clip, label="Clipped 75%")
plt.xlabel("t [s]"); plt.ylabel("V")
plt.title("Señal recortada")
plt.grid(True)


plt.figure()
plt.plot(tt, sq)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.title("Señal cuadrada de 4 kHz (s_sq)")
plt.grid(True)
plt.xlim(0, 2/fsquare)  # muestra 2 periodos
plt.show()

plt.figure()
plt.plot(tt, pulso)
plt.title("Pulso rectangular de 10 ms (s_pulse)")
plt.xlim(0, 0.03)
plt.grid(True)
```

### 2) Verificar ortogonalidad de señales

```python
ort_x1_x2 = verificar_ortogonalidad(x1, x2)
ort_x1_x3 = verificar_ortogonalidad(x1, x3)
ort_x1_sq = verificar_ortogonalidad(x1, sq)
ort_x1_pulso = verificar_ortogonalidad(x1, pulso)

print("Producto interno x1·x2 =", ort_x1_x2)
print("s1 y s2 -> ORTOGONAL")
print("Producto interno x1·x3 =", ort_x1_x3)
print("s1 y s3 -> ORTOGONAL")
print("Producto interno x1·sq =", ort_x1_sq)
print("s1 y sQ -> NO ORTOGONAL")
print("Producto interno x1·pulso =", ort_x1_pulso)
print("s1 y pulso -> ORTOGONAL")
```

### 3) Graficar la autocorrelación de la primera señal y la correlación entre esta y las demás.

```python
lags = np.arange(-len(x1) + 1, len(x1))
lags_t = lags / fs  # eje temporal (segundos)

# Calculos
auto_x1       = np.correlate(x1, x1, mode='full')
corr_x1_x2    = np.correlate(x1, x2, mode='full')
corr_x1_x3    = np.correlate(x1, x3, mode='full')
corr_x1_sq    = np.correlate(x1, sq, mode='full')
corr_x1_pulso = np.correlate(x1, pulso, mode='full')

# Eje de desplazamientos (en muestras y tiempo)
lags = np.arange(-len(x1) + 1, len(x1))
lags_t = lags / fs

# Graficos
plt.figure()
plt.plot(lags_t, auto_x1)
plt.title("Autocorrelacion de x1 (2 kHz)")
plt.xlabel("Desplazamiento [s]")
plt.grid(True)

plt.figure()
plt.plot(lags_t, corr_x1_x2)
plt.title("Correlacion entre x1 y x2")
plt.xlabel("Desplazamiento [s]")
plt.grid(True)

plt.figure()
plt.plot(lags_t, corr_x1_x3)
plt.title("Correlacion entre x1 y x3")
plt.xlabel("Desplazamiento [s]")
plt.grid(True)

plt.figure()
plt.plot(lags_t, corr_x1_sq)
plt.title("Correlacion entre x1 y senal cuadrada")
plt.xlabel("Desplazamiento [s]")
plt.grid(True)

plt.figure()
plt.plot(lags_t, corr_x1_pulso)
plt.title("Correlacion entre x1 y pulso rectangular")
plt.xlabel("Desplazamiento [s]")
plt.grid(True)

plt.show()
```

Conclusiones de la autocorrelacion.
1)Autocorrelacion: Se puede observar que la señal de autocorrelacion se encuentra sobre una envolvente y tiene un punto donde alcanza el maximo que es para s=0.Se observa que crece hasta un punto maximo y despues no se mantiene constante sino que decrece 
2)Señal desfasada: Para la correlacion para la señal desfasada se presenta el mismo comportamiento que la autocorrelacion solo que con un desfasaje.
3)Cuadrada: Para esta correlacion se observa que la señal oscila con una gran amplitud pero cuando la señal decae a 0 de la cuadrada la amplitud de la correlacion decae drasticamente.
4)Pulso: Para la correlacion cruzada de la señal con la de pulso se puede ver que para valores donde la amplitud de la cuadrada es negativa la correlacion presenta una senoidal que oscila siempre con valores positivos

### 3)Demostración trigonométrica



$$
2 \sin(\alpha)\sin(\beta) = \cos(\alpha - \beta) - \cos(\alpha + \beta) \tag{1}
$$

---

Identidades trigonométricas útiles:

$$
\cos(\alpha - \beta) = \cos(\alpha)\cos(\beta) + \sin(\alpha)\sin(\beta) \tag{2}
$$

$$
\cos(\alpha + \beta) = \cos(\alpha)\cos(\beta) - \sin(\alpha)\sin(\beta) \tag{3}
$$

---

Reemplazamos (2) y (3) en (1):

$$
2 \sin(\alpha)\sin(\beta)
= [\cos(\alpha)\cos(\beta) + \sin(\alpha)\sin(\beta)]
- [\cos(\alpha)\cos(\beta) - \sin(\alpha)\sin(\beta)] \tag{4}
$$

---

Simplificando:

$$
2 \sin(\alpha)\sin(\beta)
= 2 \sin(\alpha)\sin(\beta) \tag{5}
$$



### Demostración de la identidad con seañales senoidales.

### 4) Verificación de la identidad trigonométrica

### 2 sin(α) sin(β) = cos(α - β) - cos(α + β)
### Usando α = ωt, β = α/2

```python
f_verif = 1000.0     # Frecuencia base en Hz
fs_verif = 100_000.0 # Frecuencia de muestreo
T = 0.01             # Duración total (10 ms)
N_verif = int(fs_verif * T)
t = np.arange(N_verif) / fs_verif
omega = 2 * np.pi * f_verif

alpha = omega * t
beta = alpha / 2
# Lado izquierdo (LHS) y derecho (RHS)
lhs = 2 * np.sin(alpha) * np.sin(beta)
rhs = np.cos(alpha - beta) - np.cos(alpha + beta)

# ---- Gráficos ----
plt.figure(figsize=(9, 4))
plt.plot(t * 1e3, lhs, label="LHS = 2·sin(α)·sin(β)")
plt.plot(t * 1e3, rhs, '--', label="RHS = cos(α-β) - cos(α+β)")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("Verificación de la identidad trigonométrica (α = ωt, β = α/2)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
